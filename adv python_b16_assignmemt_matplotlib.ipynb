```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('countries.csv')
```

```python
data
```

```python
#comparing between india and china
```

```python
india=data[data.country=='India']
```

```python
india
```

```python
china=data[data.country=='China']
```

```python
china
```

```python
plt.plot(china.year,china.population/10**6)
plt.plot(india.year,india.population/10**6)
plt.xlabel('year')
plt.ylabel('population')
plt.title('china vs india population comparision')
plt.legend(['this is china','this is india'])
plt.show()
```
